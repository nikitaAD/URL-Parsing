```python
import os
import gzip
import json
import lxml.html
import dateutil
import requests
from bs4 import BeautifulSoup
try:
    # For Python 3.0 and later
    from urllib.request import urlopen
    from urllib.error import URLError
except ImportError:
    # For Python 2
    from urllib2 import urlopen

urlList = []
titleList = []
urlTitleDict = {}
unidentified = {}
stemWordsDict = {
"aboard":"preposition",
"about":"preposition",
"above":"preposition",
"across":"preposition",
"after":"preposition",
"against":"preposition",
"along":"preposition",
"amid":"preposition",
"among":"preposition",
"anti":"preposition",
"around":"preposition",
"as":"preposition",
"at":"preposition",
"before":"preposition",
"behind":"preposition",
"below":"preposition",
"beneath":"preposition",
"beside":"preposition",
"besides":"preposition",
"between":"preposition",
"beyond":"preposition",
"but":"preposition",
"by":"preposition",
"concerning":"preposition",
"considering":"preposition",
"despite":"preposition",
"down":"preposition",
"during":"preposition",
"except":"preposition",
"excepting":"preposition",
"excluding":"preposition",
"following":"preposition",
"for":"preposition",
"from":"preposition",
"in":"preposition",
"inside":"preposition",
"into":"preposition",
"like":"preposition",
"minus":"preposition",
"near":"preposition",
"of":"preposition",
"off":"preposition",
"on":"preposition",
"onto":"preposition",
"opposite":"preposition",
"outside":"preposition",
"over":"preposition",
"past":"preposition",
"per":"preposition",
"plus":"preposition",
"regarding":"preposition",
"round":"preposition",
"save":"preposition",
"since":"preposition",
"than":"preposition",
"through":"preposition",
"to":"preposition",
"toward":"preposition",
"towards":"preposition",
"under":"preposition",
"underneath":"preposition",
"unlike":"preposition",
"until":"preposition",
"up":"preposition",
"upon":"preposition",
"versus":"preposition",
"via":"preposition",
"with":"preposition",
"within":"preposition",
"without":"preposition",
"i":"pronoun",
"we":"pronoun",
"me":"pronoun",
"us":"pronoun",
"you":"pronoun",
"your":"pronoun",
"she":"pronoun",
"her":"pronoun",
"he":"pronoun",
"him":"pronoun",
"it":"pronoun",
"behind":"verb",
"says":"verb",
"did":"verb",
"how":"question",
"who":"question",
"where":"question",
"the":"determiner",
"-":"symbol",
"&":"symbol",
"|":"symbol"
}
lemmaWordsDict = {
"toaster":"object",
"2-slice":"adverb",
"compact":"adverb",
"plastic":"adverb",
"white":"adverb",
"indoorsy":"adverb",
"friend":"noun",
"man":"noun",
"woman":"noun",
"leaks":"verb",
"nsa":"noun",
"cuisinart":"objevt",
"cpt-122":"adverb"
}

urlList = ["http://blog.rei.com/camp/how-to-introduce-your-indoorsy-friend-to-the-outdoors/",
"http://www.cnn.com/2013/06/10/politics/edward-snowden-profile/",
"https://www.amazon.com/Cuisinart-CPT-122-Compact-2-Slice-Toaster/dp/B009GQ034C/ref=sr_1_1?s=kitchen&ie=UTF8&qid=1431620315&sr=1-1&keywords=toaster"]


for url in urlList:
    try:
        soup = BeautifulSoup(urlopen(url, timeout = 10),"lxml")
        title = (soup.title.string).lower()
        titleTokens = title.split()
        keywordList = []
        subjectList = []
        qualifierList = []
        for token in titleTokens:
            if(token in stemWordsDict):
                continue
            else:
                keywordList.append(token)
                if(token in lemmaWordsDict):
                    if(("object"==lemmaWordsDict[token]) or ("noun"==lemmaWordsDict[token])):
                        subjectList.append(token)
                    elif ("adverb"==lemmaWordsDict[token] or "verb"==lemmaWordsDict[token]):
                        qualifierList.append(token)
                    else:
                        unidentified[token] = "unqualified"
        if((len(subjectList) != 0) and (len(qualifierList) != 0)):
            for subject in subjectList:
                for qualifier in qualifierList:
                    phrase = qualifier + " " + subject
                    keywordList.append(phrase)
        urlTitleDict[url] = keywordList
    except URLError as e:
            print("The following HTTP Error has occurred:\n")
            print(e.code)

for entry in urlTitleDict:
    print("Url :")
    print(entry + "\n")
    print("Key Words :")
    for value in urlTitleDict[entry]:
        print(value)
    print("\n")
```
